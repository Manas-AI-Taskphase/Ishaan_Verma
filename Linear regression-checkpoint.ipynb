```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
train_file = r"C:\Users\Ishaan Verma\Downloads\train_lin.csv"
data_train= pd.read_csv(train_file)
test_file = r"C:\Users\Ishaan Verma\Downloads\test_lin.csv"
data_test = pd.read_csv(test_file)

X_train = data_train.iloc[:, :-1].values  
y_train = data_train.iloc[0:,-1:].values 

X_test = data_test.iloc[:, :-1].values  
y_test = data_test.iloc[0:,-1:].values  
print(y_train.shape)
X_train = np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test=np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
def gradient_descent(X,y,epochs,learning_rate):
    m=y.size
    theta=np.zeros((X_train.shape[1],1))
    cost_list=[]
    for i in range(epochs):    
        y_pred=np.dot(X,theta)        
        cost=(1/(2*m))*np.sum((y_pred-y)**2)
        d_theta=(1/m)*(np.dot(X.T,y_pred-y))
        theta=theta-(learning_rate*d_theta)
        cost_list.append(cost)
        if(i%100==0):
            print("Iteration: ",i," cost: ",cost)
    return theta, y_pred, cost_list
```

```python
def r_squared(y_actual, y_estimate):
    y_mean = np.mean(y_actual)
    ss_total = np.sum((y_actual - y_mean) ** 2)
    ss_residual = np.sum((y_actual - y_estimate) ** 2)
    return 1 - (ss_residual / ss_total)
```

```python
def rmse(y_actual, y_estimate):
    squared_diff = (y_actual - y_estimate) ** 2
    mean_squared_diff = np.mean(squared_diff)
    rmse_value = np.sqrt(mean_squared_diff)
    return(rmse_value)
```

```python
epochs=1000
alpha=0.0000001
coeff, y_pred_train, cost_list= gradient_descent(X_train,y_train,epochs,alpha)
print("Training accuracy: ", r_squared(y_train,y_pred_train))
print("Root mean square error: ",rmse(y_train,y_pred_train))
rng = np.arange(0,epochs)
plt.plot(rng,cost_list)
plt.show()
```

```python
#print("Intercept:", theta_best[0])
#print("Coefficients:", theta_best[1:])
y_test_pred=np.dot(X_test,coeff)
plt.scatter(X_test[:,1], y_test, label = 'Actual Value', color = 'blue')
plt.scatter(X_test[:,1], y_test_pred, label = 'Predicted Value', color = 'red')
plt.xlabel('Feature')
plt.ylabel('House Price')
plt.legend()
plt.show()
error=(1/X_test.shape[0])*np.sum(np.abs(y_test-y_test_pred))
print("Training accuracy: ", r_squared(y_test,y_test_pred))
print("Root mean square error: ",rmse(y_test,y_test_pred))
```
