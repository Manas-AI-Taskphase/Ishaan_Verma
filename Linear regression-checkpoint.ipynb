```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\Ishaan Verma\Downloads\train (1).csv'
data = pd.read_csv(file_path)
file_patt = r'C:\Users\Ishaan Verma\Downloads\test.csv'
data_test = pd.read_csv(file_patt)
X = data.iloc[:, :-1].values  
y = data.iloc[:, -1].values    
X_test = data_test.iloc[:, :-1].values  
y_test = data_test.iloc[:, -1].values  
X_b = np.c_[np.ones((X.shape[0], 1)), X]
X_b_test=np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
#Normal eqn
for index in range(10000):
    theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print("Intercept:", theta_best[0])
print("Coefficients:", theta_best[1:])
y_predict = X_b_test.dot(theta_best)
results_df = pd.DataFrame({'Actual Values': y_test, 'Predicted Values': y_predict, 'Error': y_test - y_predict})
print(results_df)
plt.scatter(y_test, y_predict)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.show()
```
